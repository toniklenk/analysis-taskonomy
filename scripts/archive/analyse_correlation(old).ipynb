# Analyse correlation

Spatial integration is the negative pearson correlation between the full image activation and the average image activation to the two halves of an image.

The spatial integration has significant spearman correlation with the beauty ratings of people.




```python
import pandas as pd
beauty_ratings = pd.read_csv('./behavior/ratings_study1.csv', header=None).mean(axis=1)
beauty_ratings.values
```

```python
integration = -1 * pd.read_csv('results_vgg16/vgg16/places1/scale8/correlations.csv', header=None)
integration
```

```python
from scipy.stats import spearmanr, pearsonr
```

```python
spearmanr(beauty_ratings.values, integration.values[:,10])
```

```python
pearsonr(beauty_ratings.values, integration.values[:,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(integration.values[:,10], beauty_ratings)
```
